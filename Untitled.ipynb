```python
import os
import pandas as pd
import numpy as np
```

```python
DIR = "New Folder 2"
df = pd.read_csv(os.path.join(DIR, "Zhang_20221117_pos_train.csv"))
df.head()
```

```python
from datetime import datetime
def time_format(time_str):
    time_obj = datetime.strptime(time_str, "%H%M%S%f")
    total_seconds = (time_obj.hour * 3600) + (time_obj.minute * 60) + time_obj.second + (time_obj.microsecond / 1e6)
    milliseconds = int(total_seconds * 1000)
    
    return milliseconds
```

```python
img2_label_values = df['start,end,truth'].values
image1s, image2s, image3s, labels, time2_in_mss = [], [], [], [], []
for img2_label_value in img2_label_values:
    image1, image3, label = img2_label_value.split(',')
    image_prefix = image1.rsplit('_', 1)[0] + '_'
    time1 = int(image1.rsplit('_', 1)[1].split('.')[0])
    time3 = int(image3.rsplit('_', 1)[1].split('.')[0])
    time2_str = f"{(time1+time3) // 2:09d}"
    time2_in_mss.append(time_format(time2_str))
    
    image2s.append(image_prefix + time2_str + '.png')
    image1s.append(image1)
    image3s.append(image3)
    labels.append(1)

df_new = pd.DataFrame({
    'image1': image1s,
    'image2': image2s,
    'image3': image3s,
    'label' : labels
})
df_new.to_csv('new.csv')
df_new.head()
```

```python
# 在df_new 的基础上，继续生成新的pairs，包含opposite的pairs
import random

new_rows = []
PREX = "RawData/1_Zhang_20221117_F/Trip_image_aligned_cut/Cropped_"
for index, row in df_new.iterrows():
    img1, img2, img3 = PREX+row['image1'], PREX+row['image2'], PREX+row['image3']
    # change direction: same
    new_rows.append({'image_a': img1, 'image_b': img2, 'image_c': img3, 'label': 1})
    new_rows.append({'image_a': img3, 'image_b': img2, 'image_c': img1, 'label': 1})
    
    # change direction: opposite
    opposite_rows = [
        {'image_a': img1, 'image_b': img3, 'image_c': img2, 'label': 0},
        {'image_a': img2, 'image_b': img1, 'image_c': img3, 'label': 0},
        {'image_a': img2, 'image_b': img3, 'image_c': img1, 'label': 0},
        {'image_a': img3, 'image_b': img1, 'image_c': img2, 'label': 0}
    ]
    # delete 2 rows in opposite randomly
    rows_to_keep = random.sample(opposite_rows, 2)
    new_rows.extend(rows_to_keep)

    # break
df_4 = pd.DataFrame(new_rows)
df_4.to_csv("Zhang_20221117_pos_train_trip.csv", index=False)
```

```python
from PIL import Image

for index, row in df_new.iterrows():
    # 读取每行的三个图片路径
    img1_path, img2_path, img3_path = row['image1'], row['image2'], row['image3']
    
    # 使用 PIL 打开图片
    img1 = Image.open(img1_path)
    img2 = Image.open(img2_path)
    img3 = Image.open(img3_path)
    
    # 获取每个图片的宽高
    width1, height1 = img1.size
    width2, height2 = img2.size
    width3, height3 = img3.size
    
    # 创建一个新的图像，其宽度是三张图片的总宽度，高度是它们中最大值
    total_width = width1 + width2 + width3
    max_height = max(height1, height2, height3)
    
    # 创建一个新的空白图片来存放拼接结果
    new_img = Image.new('RGB', (total_width, max_height))
    
    # 将三张图片依次粘贴到新图像上
    new_img.paste(img1, (0, 0))  # 粘贴第一张图片
    new_img.paste(img2, (width1, 0))  # 粘贴第二张图片
    new_img.paste(img3, (width1 + width2, 0))  # 粘贴第三张图片
    
    # 选择保存或显示拼接后的图片
    # new_img.show()  # 显示拼接后的图片
    new_img.save(f'output_image_{index}.png')
```

```python
# 使用 OpenCV 从视频文件中提取对应帧
import cv2
video_file = "path_to_your_video.mp4"
cap = cv2.VideoCapture(video_file)

fps = cap.get(cv2.CAP_PROP_FPS)
for time2_in_ms in time2_in_mss:
    frame_number = int((time2_in_ms / 1000) * fps)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    ret, frame = cap.read()
    
    if ret:
        cv2.imwrite()
    else:
        print("无法读取指定的帧")
cap.release()
```

## Py-Feat 提取AU强度和表情强度

```python
import cv2
import os

DIR = "Nagoya Elder Facial Expression Dataset (No001)"
video_path = os.path.join(DIR, "No001_m_fpi_01hapo_f.mp4")
cap = cv2.VideoCapture(video_path)

frames = []
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)

cap.release()
```

```python
from feat import Detector

# 初始化 py-Feat 检测器
detector = Detector()

# 对每一帧进行表情和AU分析
results = []
for frame in frames:
    result = detector.detect_image(frame)
    results.append(result)
```

```python
import pandas as pd

# 创建一个用于存储结果的 DataFrame
au_data = pd.DataFrame()

for result in results:
    au_data = au_data.append(result.aus, ignore_index=True)

# 查看AU强度结果
print(au_data)
```

```python
au_data.to_csv("au_intensity_results.csv", index=False)
```
